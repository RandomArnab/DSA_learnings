### Largest element in an array.

```python
def max_ele(nums):
    res = nums[0]
    for num in nums:
        res = max(num, res)
    return res

print(max_ele([2, 5, 1, 3, 0]))
```

### Second Largest Element in array.

```python
def max_ele2(nums):
    res = nums[0]
    max_ele = max(nums)
    for num in nums:
        if (num>res) and (num!=max_ele):
            res = max(res,num)
    return res

print(max_ele2([1, 2, 4, 7, 6, 5]))
```

### Check if array is sorted and rotated.

```python
def check_arr(nums):
    res = 0
    for i in range(len(nums)):
        if nums[i]>nums[(i+1)%len(nums)]:
            res += 1
    return res<=1

print(check_arr([3,4,5,1,2]))
print(check_arr([2,1,3,4]))
```

### Remove duplicates from sorted array.

```python
def rem_dup(nums):
    dicted = {}
    new_arr = []
    for num in nums:
        if num in dicted.keys():
            dicted[num] += 1
        else:
            dicted[num] = 1
    new_arr = [v for v in dicted.keys()]
    new_arr.extend(['_' for _ in range(len(nums)-len(new_arr))])
    return new_arr

print(rem_dup([1,1,2]))
print(rem_dup([0,0,1,1,1,2,2,3,3,4]))        
```

### Left Rotate array by k places.

```python
def left_rotate_k(nums,k):
    k = k%len(nums)
    nums.reverse()
    nums[:k] = reversed(nums[:k])
    nums[k:] = reversed(nums[k:])
    return nums

print(left_rotate_k([1,2,3,4,5,6,7],3))
print(left_rotate_k([-1,-100,3,99],2))
```

### Move Zeroes to end.

```python
def move_zeroes(nums):
    j=0
    for i in range(len(nums)):
        if nums[i]==0:
            nums[i],nums[j] = nums[j], nums[i]
            j += 1
    return nums

print(move_zeroes([1,3,12,0,0]))
print(move_zeroes([0]))
```

## Linear Search.

```python
def linear_search(nums,k):
    for i in range(len(nums)):
        if nums[i] == k:
            return i
    return -1

print(linear_search([1,2,3,4,5],3))
```

### Union of two sorted arrays.

```python
def union_sorted(arr1, arr2):
    i,j = 0,0
    res = []

    while i < len(arr1) and j < len(arr2):

        if i > 0 and arr1[i] == arr1[i - 1]:
            i += 1
            continue

        if j > 0 and arr2[j] == arr2[j - 1]:
            j += 1
            continue

        if arr1[i] < arr2[j]:
            res.append(arr1[i])
            i += 1
        elif arr1[i] > arr2[j]:
            res.append(arr2[j])
            j += 1
        else:
            res.append(arr1[i])
            i += 1
            j += 1

    while i < len(arr1):
        if i == 0 or arr1[i] != arr1[i - 1]:
            res.append(arr1[i])
        i += 1

    while j < len(arr2):
        if j == 0 or arr2[j] != arr2[j - 1]:
            res.append(arr2[j])
        j += 1

    return res

arr1 = [1, 2, 2, 3, 4]
arr2 = [2, 3, 5, 5, 6]
print(union_sorted(arr1, arr2))
```

### Find missing number.

```python
def miss_num(nums):
    n = len(nums)+1
    return (n*(n+1)//2) - sum(nums)

print(miss_num([1,2,4,5,6,7,8]))
```

### Maximum Consecutive Ones.

```python
def max_ones(nums):
    max_ones, curr_ones = 0,0
    for i in range(len(nums)):
        if nums[i]==1:
            curr_ones += 1
            max_ones = max(max_ones,curr_ones)
        else:
            curr_ones = 0
    return max_ones

print(max_ones([1,1,0,1,1,1]))
```

### Find the number that appears once, and the other numbers twice.

```python
def find_single_num(nums):
    dicted = {}
    for num in nums:
        if num in dicted.keys():
            dicted[num] += 1
        else:
            dicted[num] = 1
    return [k for k,v in dicted.items() if v==1][0]

print(find_single_num([2,2,3,3,1]))
```

### Longest Subarray with given Sum K(Positives).

```python
def longest_subarray_pos(nums, k):
    l = 0
    curr_sum,max_len = 0,0
    for r in range(len(nums)):
        curr_sum += nums[r]
        while l<=r and curr_sum>k:
            curr_sum -= nums[l]
            l += 1
        if curr_sum == k:
            max_len = max(max_len, r-l+1)
    return max_len

print(longest_subarray_pos([2,3,5,1,9],10))
print(longest_subarray_pos([10, 5, 2, 7, 1, 9],15))

```

### Length of the longest subarray with zero Sum.

```python
def longest_subarray_zero_sum(nums):
    prefix_sum, max_len = 0,0
    seen = {0:-1}
    for i in range(len(nums)):
        prefix_sum += nums[i]
        if prefix_sum in seen:
            max_len = max(max_len, i - seen[prefix_sum])
        else:
            seen[prefix_sum] = i
    return max_len

print(longest_subarray_zero_sum([1, -1, 3, 2, -2, -3, 3]))
print(longest_subarray_zero_sum([1, 2, -3, 3]))            
print(longest_subarray_zero_sum([1, 2, 3]))    
```
