### Bubble Sort.

```python
def bubble_sort(nums):

    for i in range(len(nums)):
        for j in range(len(nums)):
            if nums[j] > nums[i]:
                nums[i], nums[j] = nums[j], nums[i]
    return nums

print(bubble_sort([40,10,20,30,50]))
```

### Selection Sort.

```python
def selection_sort(nums):
    for i in range(len(nums)):
        min_index = i
        for j in range(i+1,len(nums)):
            if nums[j]<nums[min_index]:
                nums[j], nums[min_index] = nums[min_index], nums[j]
    return nums

print(selection_sort([10,14,27,33,35,19,42,44]))
```

## Insertion Sort.

```python
def insertion_sort(nums):
    for index in range(len(nums)):
        value = nums[index]
        i = index-1
        while (i>=0) & (value<nums[i]):
            nums[i],nums[i+1] = nums[i+1],nums[i]
            i -= 1
    return nums

print(insertion_sort([[10,14,27,33,35,19,42,44]]))
```

## Merge Sort.

```python
def merge_sort(nums):

    if len(nums)<=1:
        return nums

    # divide.
    left_arr = nums[:len(nums)//2]
    right_arr = nums[len(nums)//2:]

    # recursion.
    merge_sort(left_arr)
    merge_sort(right_arr)

    # merge.
    i,j,k = 0,0,0
    while (i<len(left_arr)) & (j<len(right_arr)):
        if left_arr[i]<right_arr[j]:
            nums[k] = left_arr[i]
            i+=1
        else:
            nums[k] = right_arr[j]
            j+=1
        k+=1
    
    while i<len(left_arr):
        nums[k] = left_arr[i]
        i+=1
        k+=1
    while j<len(right_arr):
        nums[k] = right_arr[j]
        j+=1
        k+=1

    return nums

print(merge_sort([10,14,27,33,35,19,42,44]))




```

## Quick Sort.

```python
def quick_sort(nums):
    if len(nums)<=1:
        return nums
    pivot = nums[-1]
    left = [x for x in nums[:-1] if x<pivot]
    right = [x for x in nums[:-1] if x>pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)

print(quick_sort([22, 11, 88, 66, 55, 77, 33,44]))
```

## Bucket Sort.

```python
def bucket_sort(nums):

    if len(nums)<=1:
        return nums
    
    min_val, max_val = min(nums), max(nums)
    bucket_range = int((max_val - min_val)/len(nums))
    buckets = [[] for _ in range(bucket_range)]

    for num in nums:
        index = int((num - min_val)/bucket_range)
        buckets[index].append(num)
    
    for bucket in buckets:
        bucket.sort()

    res = []
    for bucket in buckets:
        res.extend(bucket)
    
    return res

print(bucket_sort([22,11,88,66,55,77,33,44]))
```
